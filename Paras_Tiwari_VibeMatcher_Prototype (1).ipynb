```python
from langchain_community.document_loaders import JSONLoader

loader = JSONLoader(
    file_path="data.json",
    jq_schema=".[] | {name: .name, desc: .desc, tags: .tags, category: .category, price_range: .price_range}",
    text_content=False
)
data = loader.load()
data[0]
```

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['hugging_face_api']=os.getenv("hugging_face_api")
```

```python
from langchain_community.embeddings import HuggingFaceEmbeddings
embedding=HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")
```

```python
texts = [doc.page_content for doc in data]
result = embedding.embed_documents(texts)
```

```python
result
```

```python
from langchain_community.vectorstores import FAISS
vectorstore = FAISS.from_documents(data, embedding)
```

```python
groq_api_key=os.getenv("GROQ_API_KEY")
```

```python
pip install langchain_groq
```

```python
from langchain_groq import ChatGroq

llm = ChatGroq(model_name='openai/gpt-oss-120b', groq_api_key=groq_api_key)
```

```python
from langchain_core.prompts import ChatPromptTemplate
generic_template="""You are a **fashion vibe recommender system**.
Your task is to recommend **the top 3 fashion products** that best match the user's vibe query.
You are given product descriptions, tags, and categories as context.
Carefully analyze the vibe-related keywords (e.g., cozy, urban, chic, sporty, elegant, traditional, etc.)
and match them semantically to the user's requested vibe. """
prompt=ChatPromptTemplate.from_messages(
    [
        "system",generic_template,
        "user","{text}"
    ]
)
```

```python
retriever = vectorstore.as_retriever()
```

```python
!pip install -U langchain langchain-core langchain-community langchain-groq
```

```python
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
chain = (
    {"context": retriever, "text": RunnablePassthrough()}
    | prompt
    | llm
    | StrOutputParser()
)
from IPython.display import Markdown, display

result = chain.invoke("Suggest three outfit options that fit an energetic and urban streetwear vibe for winter.")
display(Markdown(result))

```
